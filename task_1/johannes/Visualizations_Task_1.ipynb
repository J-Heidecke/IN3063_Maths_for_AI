```python
# Data
# Results of the baseline algorithm
baseline_results = {'10': 87 ,
                   '20' : 168,
                   '30': 271,
                   '40' : 357,
                   '50': 398,
                   '60' : 525,
                   '70': 576,
                   '80' : 684 ,
                   '90': 784,
                   '100' : 884 }

# Results of the dijkstra algorithm
dijkstra_results = {'10': 50,
                   '20' : 84 ,
                   '30': 130 ,
                   '40' : 172}

# Results of the ant-colony algorithm
ant_results = {'10': 83,
                   '20' : 198}

# Time taken by the baseline algorithm
baseline_time = {'10': 0.0009968280792236328 ,
                   '20' : 0.00099945068359375 ,
                   '30': 0.0009975433349609375,
                   '40' : 0.000997781753540039,
                   '50': 0.00099945068359375,
                   '60' : 0.0009975433349609375,
                   '70': 0.000997304916381836,
                   '80' : 0.000998860824584961,
                   '90': 0.0009989738464355469,
                   '100' : 0.0009999275207519531}

# Time taken by the dijkstra algorithm
dijkstra_time = {'10': 0.011944293975830078 ,
                   '20' : 0.39513230323791504,
                   '30': 4.022212743759155,
                   '40' : 22.54074740409851 
                }

# Time taken by the any-colony algorithm
ant_time =  {'10': 1.1983134746551514,
                   '20' : 2.95731520652771}
```

## Baseline

```python
# Baseline Results Visualization
import matplotlib.pyplot as plt
import numpy as np

plt.plot(baseline_results.keys(), baseline_results.values())
plt.xlabel('Gride Size / N x N')
plt.ylabel('Length of Path')
```

```python
# Basline Time Visualization 
plt.plot(baseline_time.keys(), baseline_time.values())
plt.xlabel('Gride Size / N x N')
plt.ylabel('Time / s')
 
print(np.mean(list(baseline_time.values())))
```

```python
# Combined 
colour1 = '#F0A941'
colour2 = '#0B62A3'

fig,ax = plt.subplots()
ax.plot(baseline_results.keys(), baseline_results.values(), color=colour1)
ax.set_xlabel('Gride Size / N x N')
ax.set_ylabel('Length of Path',color=colour1)
ax2=ax.twinx()
ax2.plot(baseline_time.keys(), baseline_time.values(),color=colour2, alpha = 0.4)
ax2.set_ylabel('Time / s',color=colour2)
plt.show()
```
